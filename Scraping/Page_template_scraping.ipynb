```python
titles = []
text_content = []

# Find all the <span> elements with a specific class
spans_with_class = soup.find_all('span', class_= 'wysiwyg-underline')  # Replace with your specific class

for i in range(len(spans_with_class) - 1):
    start_span = spans_with_class[i]
    end_span = spans_with_class[i + 1]

    # Extract the text between the current and next span
    text_between_spans = ' '.join([str(tag) for tag in start_span.find_next_siblings() if tag != end_span])

    titles.append(start_span.text.strip())
    text_content.append(text_between_spans.strip())

```
